### Проверка расчета эффективности обогащения

Вычислю её на обучающей выборке по формуле, потом проверю в сравнении с целевым признаком.

```python
# вычисление коэффициента Recovery по формуле: Recovery = (C * (F - T)) / (F * (C - T))
F = data_train_del['rougher.input.feed_au']
C = data_train_del['rougher.output.concentrate_au']
T = data_train_del['rougher.output.tail_au']
recovery = (C * (F - T)) / (F * (C - T)) * 100
recovery
```

Найдем MAE между Recovery и целевым значением признака из target_train.

```python
# MAE: воспользуемся встроенной в библиотеку функцией вычисления среднего абсолютного отклонения
MAE = mean_absolute_error(target_train['rougher.output.recovery'], recovery)
print(f'MAE = {MAE}')
print(f'Среднее арифметическое целевого признака = {target_train["rougher.output.recovery"].mean()}')
print(f'Процент МАЕ от среднего значения целевого признака = {MAE / target_train["rougher.output.recovery"].mean() * 100}')
```

**Вывод:**  
Среднее асболютное отклонение составляет **менее 0,1%**. Отсюда выводим, что данные обучающей выборки корректны, отклонение имеет очень маленькую величину.

### Функция для вычисления итоговой sMAPE

```python
def smape(target, predict):
    """считает метрику sMAPE- Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное 
    отклонение»по формуле: sMape = 1/N * sum((abs(y[i] - y1[i]) / (abs(y[i]) + abs(y1[i])) / 2) * 100%) for i in range(1, N + 1))
    где y - target, y1 - predict"""
    N = len(target)
    target = pd.Series(target)
    target_index = target.index
    predict = pd.Series(predict, index=target_index)
    error = (target - predict).abs()
    scale = ((target).abs() + (predict).abs()) / 2  
    return ((1 / N) * (error / scale) * 100).mean()
```

```python
def end_smape(rougher, final):
    """считает итоговую метрику sMAPE по формуле: итоговое sMape = 0.25 * sMAPE(rougher) + 0.75 * sMAPE(final)"""
    return 0.25 * rougher + 0.75 * final
```

```python
def smape(target_1, predictions):    
    a = abs(target_1 - predictions)
    b = (abs(target_1) + abs(predictions)) / 2
    errors = a / b
    return errors.sum()/ errors.shape[0] * 100

def ismape(smape_rougher, smape_final):
    result = 0.25*smape_rougher + 0.75*smape_final
    return result
```
