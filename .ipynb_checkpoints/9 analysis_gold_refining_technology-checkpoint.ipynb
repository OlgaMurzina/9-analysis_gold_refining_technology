### Проверка расчета эффективности обогащения

Вычислю её на обучающей выборке по формуле, потом проверю в сравнении с целевым признаком.

```python
# вычисление коэффициента Recovery по формуле: Recovery = (C * (F - T)) / (F * (C - T))
F = data_train_del['rougher.input.feed_au']
C = data_train_del['rougher.output.concentrate_au']
T = data_train_del['rougher.output.tail_au']
recovery = (C * (F - T)) / (F * (C - T)) * 100
recovery
```

Найдем MAE между Recovery и целевым значением признака из target_train.

```python
# MAE: воспользуемся встроенной в библиотеку функцией вычисления среднего абсолютного отклонения
MAE = mean_absolute_error(target_train['rougher.output.recovery'], recovery)
print(f'MAE = {MAE}')
print(f'Среднее арифметическое целевого признака = {target_train["rougher.output.recovery"].mean()}')
print(f'Процент МАЕ от среднего значения целевого признака = {MAE / target_train["rougher.output.recovery"].mean() * 100}')
```

**Вывод:**  
Среднее асболютное отклонение составляет **менее 0,1%**. Отсюда выводим, что данные обучающей выборки корректны, отклонение имеет очень маленькую величину.

## Модель

### Функция для вычисления итоговой sMAPE

```python
def smape(target_1, predictions): 
    """считает метрику sMAPE- Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное 
    отклонение»по формуле: sMape = 1/N * sum((abs(y[i] - y1[i]) / (abs(y[i]) + abs(y1[i])) / 2) * 100%) for i in range(1, N + 1))
    где y - target, y1 - predict"""
    a = abs(target_1 - predictions)
    b = (abs(target_1) + abs(predictions)) / 2
    errors = a / b
    return errors.sum()/ errors.shape[0] * 100
```

```python
def ismape(smape_rougher, smape_final):
    """считает итоговую метрику sMAPE по формуле: итоговое sMape = 0.25 * sMAPE(rougher) + 0.75 * sMAPE(final)"""
    result = 0.25*smape_rougher + 0.75*smape_final
    return result
```

### Обучение разных моделей и оценка их качества кросс-валидацией

Имеется несколько линейных моделей для предсказания нескольких значений непрерывной величины, в которых сразу примененена оценка качества кросс-валидацией. Перечислю те, которые будут использованы в этом исследовании:
1) LinearRegression()  
2) Lasso()  
3) ExtraTreesRegressor()  
4) RandomForestRegressor()  
5) DecisionTreeRegressor()  

Будем запускать эти модели в цикле на обучение, фиксировать лучший результат обучения на обучающей выборке, его и будем проверять на тестовой выборке в дальнейшем. 

```python
target_rougher_train = target_train['rougher.output.recovery']
target_final_train = target_train['final.output.recovery']
target_rougher_train, target_final_train
```

Проведем обучение наших моделей (настройки по умолчанию) на подготовленных данных, сделаем выводы.

```python
random_state = 12345
cv = 5
models = [DecisionTreeRegressor(random_state = random_state), 
          RandomForestRegressor(random_state=random_state), 
          LinearRegression(), 
          Lasso(), 
          ExtraTreesRegressor(random_state=random_state)]


results_cross_val = []
best_res = 100
best_model = None

for model in models: 
    
    scorer = make_scorer(smape, greater_is_better=False) 
    
    cross_val_score_rougher = cross_val_score(model, 
                                              features_train, 
                                              target_rougher_train, 
                                              cv=cv, scoring=scorer).mean()
    cross_val_score_final = cross_val_score(model, 
                                            features_train, 
                                            target_final_train, 
                                            cv=cv, scoring=scorer).mean()

    results_cross_val.append({'model name': model.__class__.__name__, 
                              'cross_val_score_rougher': cross_val_score_rougher, 
                              'cross_val_score_final': cross_val_score_final}) 
    esmape = abs(ismape(cross_val_score_rougher, cross_val_score_final))
    if esmape < best_res:
        best_res = esmape
        best_model = model
              
    print(f'Модель {model} дает метрику Итоговое sMAPE = {esmape}')
print()
print(f'Лучшая модель - {best_model} с метрикой Итоговое sMAPE = {best_res}')
```
